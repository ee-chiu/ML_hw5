```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import json
pd.options.mode.chained_assignment = None  # default='warn'

from collections import OrderedDict
from operator import itemgetter    
```

## 1. Data Input

```python
def get_train(train):
    #train = train[['id', 'ingredients', 'cuisine']]
    size_list = []
    for i in range(len(train['ingredients'])):
        size_list.append(len(train['ingredients'][i]))
        train['ingredients'][i] = ",".join(train['ingredients'][i])
        
    split = train['ingredients'].str.split(',', expand = True)
    train = pd.concat([train, split], axis = 1)
    train['size'] = size_list
    train = train.drop('ingredients', axis = 1)
    
    X_train = train.drop('cuisine', axis = 1)
    y_train = train['cuisine']
    
    X_train = pd.get_dummies(X_train, prefix = "", prefix_sep = "")
    
    return X_train, y_train
```

```python
def get_test(test):
    #test = test[['id', 'ingredients']]
    for i in range(len(test['ingredients'])):
        test['ingredients'][i] = ",".join(test['ingredients'][i])
    split = test['ingredients'].str.split(',', expand = True)
    test = pd.concat([test, split], axis = 1)
    test = test.drop('ingredients', axis = 1)
    
    X_test = pd.get_dummies(test, prefix = "", prefix_sep = "")
    
    return X_test
```

```python
def get_dict(train):
    count_dict = {}
    
    for i_list in train['ingredients']:
        for i in i_list:
            if i not in count_dict:
                count_dict[i] = 0
            count_dict[i] += 1
            
    order_dict = OrderedDict(sorted(count_dict.items(), key = itemgetter(1), reverse = True))
            
    return order_dict
```

```python
def load_data():
    with open("train.json") as train_data:
        train = json.load(train_data)
    with open("test.json") as test_data:
        test = json.load(test_data)
    
    train = pd.DataFrame(train)
    order_dict = get_dict(train)
    X_train, y_train = get_train(train)
    
    test = pd.DataFrame(test)
    X_test = get_test(test)
    
    return X_train, y_train, X_test, order_dict
```

```python
X_train, y_train, X_test, order_dict = load_data()
```

## 2. Data Visualization

* Plot the data distribution by value count of their labels (targets).

```python
def draw1(y_train):
    indices = y_train.value_counts().index
    values = y_train.value_counts().values
    
    plt.figure(figsize = (20, 16), dpi = 100)
    ax = sns.barplot(x = indices, y = values)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")

    plt.show()
    
    return
```

```python
draw1(y_train)
```

```python
def draw2(order_dict):
    order_list = list(order_dict.items())
    indices = []
    values = []
    
    
    for i in range(30):
        indices.append(order_list[i][0])
        values.append(order_list[i][1])
    
    plt.figure(figsize = (20, 16), dpi = 100)
    ax = sns.barplot(x = indices, y = values)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")

    plt.show()
    
    return
```

```python
draw2(order_dict)
```
